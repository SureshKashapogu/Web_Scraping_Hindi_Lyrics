We are going to scrape Hindi Lyrics from multiple pages from a website and store it as a txt file

```python
# import the libraries

from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://songsofshailendra.com/2013/08/27/shailendra-all-songs-listing/'
r = requests.get(url)
print(r)
# response code 200 means that scraping is allowed from the website
# response code 403 means that scraping is NOT allowed from the website
```

```python
hrefs = []
# retrieve the data from URL
response = requests.get(url)
if response.status_code ==200:
    soup=BeautifulSoup(response.text, 'html.parser')
    for tag in soup.find_all(href=True):
        hrefs.append(tag['href'])
    
print(hrefs)
```

```python
df = pd.DataFrame(hrefs) 
    
# saving the dataframe 
df.to_csv('hrefs.csv')
```
